```python
from sklearn.preprocessing import MinMaxScaler
trans = MinMaxScaler()
data=pd.DataFrame(trans.fit_transform(Airline.iloc[:,1:]))
data
```

HIERARCHICAL CLUSTERING

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
z=linkage(data,method='average',metric='euclidean')
plt.figure(figsize=(15,5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
h_complete=AgglomerativeClustering(n_clusters=5,linkage='average',affinity='euclidean').fit(data)
cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
```

```python
Airline['HCluster']=cluster_labels
Airline
```

K MEANS CLUSTERING

```python
from sklearn.cluster import KMeans
fig =plt.figure(figsize=(10,8))
WCSS=[]
for i in range(1,11):
  clf=KMeans(n_clusters=i)
  clf.fit(data)
  WCSS.append(clf.inertia_)
plt.plot(range(1,11),WCSS)
plt.title("The Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
clf=KMeans(n_clusters=4)
y_kmeans=clf.fit_predict(data)
```

```python
clf.labels_
```

```python
y_kmeans
```

```python
clf.cluster_centers_
```

```python
clf.inertia_
```

```python
md=pd.Series(y_kmeans)
Airline['KCluster']=md
Airline
```

```python
Airline.iloc[:,1:].groupby(Airline.KCluster).mean()
```

```python
plt.style.use(['classic'])
Airline.plot(x="Balance",y="Qual_miles",c=clf.labels_,kind='scatter',s=30,cmap=plt.cm.coolwarm)
plt.title('Clusters using K Means')
```

DBSCAN CLUSTERING

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan=DBSCAN(eps=0.475,min_samples=5)
dbscan.fit(data)
```

```python
dbscan.labels_
```

```python
db=pd.DataFrame(dbscan.labels_,columns=['DBCluster'])
db
```

```python
Airline['DBCluster']=db
Airline
```

```python
df=Airline.drop(columns=['HCluster','KCluster'])
df
```

```python
df.plot(x="Balance",y="Qual_miles",c=dbscan.labels_,kind='scatter',s=30,cmap=plt.cm.copper_r)
plt.title("Clusters using DBScan")
```
