IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
```

EXPLORATION

```python
df=pd.read_csv('book.csv',encoding='latin-1')
df
```

```python
df2=df.drop('Unnamed: 0',axis=1) #DROPPING UNNAMED COLUMN
df2
```

```python
books_df=df2.sort_values('User.ID').reset_index(drop='True')
#SORTING VALUES IN ASCENDING ORDER OF USERID AND RESETTING INDEX
books_df
```

```python
len(books_df['User.ID'].unique())
```

```python
len(books_df['Book.Title'].unique())
```

```python
user_books_df =books_df.pivot_table(index='User.ID',
                                 columns='Book.Title',
                                 values='Book.Rating').reset_index()3CREATING PIVOT TABLE
user_books_df
```

```python
user_books_df.index=books_df['User.ID'].unique()#SETTING INDEX AS UNIQUE USER ID'S
user_books_df
```

```python
user_books_df.fillna(0, inplace=True)#FILLING NULL VALUES WITH 0
user_books_df
```

FINDING COSINE SIMILARITY

```python
from sklearn.metrics import pairwise_distances
```

```python
user_sim = 1 - pairwise_distances( user_books_df.values,metric='cosine')
user_sim
```

```python
user_sim_df=pd.DataFrame(user_sim)
user_sim_df
```

```python
user_sim_df.index=books_df['User.ID'].unique()
user_sim_df.columns=books_df['User.ID'].unique()
user_sim_df
```

```python
user_sim_df.iloc[0:5,0:5]
```

```python
np.fill_diagonal(user_sim,0)# FILLING NULL VALUES WITH 0
user_sim_df.iloc[0:5,0:5]
```

```python
user_sim_df.idxmax(axis=1)[0:5]#FINDING SIMILAR USERS
```
