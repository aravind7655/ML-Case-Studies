```python
wine_data = wine.drop(wine.columns[0], axis=1)#dropping first column as it shows clusters and not required
wine_data
```

PREPROCESSING DATA

```python
wine_normal= scale(wine_data)#Normalising data
```

```python
wine_normal
```

PCA

```python
pca=PCA(n_components=6)
pca_values=pca.fit_transform(wine_normal)
pca_values
```

```python
pca.components_
```

```python
var =pca.explained_variance_ratio_
var
```

```python
var1=np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
plt.plot(var1,color="red")
```

```python
final=pd.DataFrame(columns=["pc1","pc2","pc3"])
```

```python
final["pc1"]=pd.Series(pca_values[:,0])
final["pc2"]=pd.Series(pca_values[:,1])
final["pc3"]=pd.Series(pca_values[:,2])
```

```python
final
```

HIERARCHICAL CLUSTERING

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch
z =linkage(final,method='average',metric='euclidean')
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
import warnings
warnings.filterwarnings('ignore')
h_complete=AgglomerativeClustering(n_clusters=3,linkage="average",affinity="euclidean").fit(final)
cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
final["clust"]=cluster_labels
```
