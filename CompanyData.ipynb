```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn import preprocessing
```

```python
company = pd.read_csv('https://excelr.s3.nl-ams.scw.cloud/DataScience-Assignments/Assignments/Decision%20Tree/Company_Data.csv')
company
```

```python
label_encoder=preprocessing.LabelEncoder()
company["ShelveLoc"]=label_encoder.fit_transform(company["ShelveLoc"])
company["Urban"]=label_encoder.fit_transform(company["Urban"])
company["US"]=label_encoder.fit_transform(company["US"])
```

```python
company
```

```python
company["Sales"].mean()
```

```python
company["Sales"]=np.where(company.Sales < 7.49,"small","large")
company
```

```python
x=company.iloc[:,1:]
y=company.Sales
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify = y)
```

```python
model = DecisionTreeClassifier(criterion = 'entropy',max_depth=3)
model.fit(x,y)
```

```python
tree.plot_tree(model);
```

```python
model.feature_importances_
```

```python
fi = pd.DataFrame({'feature': list(x_train.columns),
                   'importance': model.feature_importances_}).\
                    sort_values('importance', ascending = False)
```

```python
fi
```

```python
model_gini = DecisionTreeClassifier(criterion='gini', max_depth=3)
```

```python
model_gini.fit(x_train, y_train)
```

```python
model_gini.feature_importances_
```
